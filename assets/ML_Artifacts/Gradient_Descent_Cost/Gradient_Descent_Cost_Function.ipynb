## Calculating cost with gradient descent and learning rate
- Change the iteration and learning rate vaules and see the impact on cost.
- Low iteration values with high learning rate (i.e. big steps) may lead to miss the global minimum
- Goal is to reach minimum cost with minimum iteration

gradient.png

```python
# code credit:codebasics https://codebasics.io/coming-soon

import numpy as np

def gradient_descent(x,y):
    m_curr = b_curr = 0
    #iterations = 100       #change value
    iterations = 100       #change value
    n = len(x)
    #learning_rate = 0.08   #change value
    learning_rate = 0.16   #change value

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```

**Insights**:

This code snippet defines a function gradient_descendent (x, y) which implements the gradient descent algorithm. The function initializes the slope m and y-intercept b of the linear model to zero. It then, iterates a specified number of times (default 100 iterations), adjustiung m and b to minimize the cost function, which is the mean squared error between the predicted y-values and the actual y-values. The learning rate is set to 0.08. It prints the values of m, b, and the cost at each iteration.

* The gradient descent algorithm, run with its default parameters (100 iterations and a learning rate of 0.08), resulted in a final cost of approximately **0.0041** after the 100th iteration.

* Increased Iterations (200 iterations, learning rate 0.08): The final cost is approximately **1.711074266712627e-05**. This is lower than the default run, indicating that increasing the number of iterations allows the algorithm to further minimize the cost.

* Decreased Learning Rate (100 iterations, learning rate 0.04): The final cost is approximately **0.0634**. This is significantly higher than the default run, suggesting that a lower learning rate may not allow the algorithm to converge to the minimum cost within the same number of iterations.

* Increased Learning Rate (100 iterations, learning rate 0.16): The final cost is approximately **1.1171143851204643e+90** . This extremely high value indicates that the cost function diverged, likely due to the learning rate being too high.

**We can coclude the following**:

Increasing the Number of Iterations: Generally allows the algorithm more opportunity to converge towards the minimum cost. This is beneficial but also increases computation time.

Decreasing the Learning Rate: May require more iterations to converge and can lead to underfitting if the rate is too low or the number of iterations isn't sufficiently high.

Increasing the Learning Rate: Can cause the algorithm to overshoot the minimum and diverge, as seen in the extremely high cost in the third experiment. A too high learning rate is detrimental to the convergence of the algorithm.
